```python
import numpy as np
# colummn : Age ,math marks,science mark
data=np.array([
    [18,85,78], # student 1
    [19,92,88], # student 2 # 1s column 
    [20,78,65], # student 3
    [19,76,89], # student 4
    [12,87,96]  # student 5
])
```

```python
# get the shape of the matrix 
data.shape
```

```python
# find the average age of student 
np.mean(data[:,0])
```

```python
# extracts maths marks of all student
data[:,1]
```

```python
# find the highest science marks
np.max(data[:,2])
```

```python
# Get the detail pf the student who scored more than 90 in maths 
# using boolean
data[data[:,1] > 90]

```

```python
# increase maths marks by 5 
data[:,1]+=5
data
```

```python
# find how many stundent is younger than 19 
len(data[data[:,0]<19])
```

```python
# calculate the average marks in each subject 
np.mean(data[:,1:],axis=0)
```

```python
# get data of student at least 80 i two subject
data[(data[:,1]>=80) & (data[:,2])>=80]
```

```python
# replace all science marks <75 with 0
data[:,2][data[:,2]<75]= 0
data
```
